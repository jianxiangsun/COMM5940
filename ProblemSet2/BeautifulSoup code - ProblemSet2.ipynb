```python
%%time
import requests
import csv
import pandas as pd
from bs4 import BeautifulSoup
from time import sleep
from random import randint
from time import time

header = ['page #','job_title','url','companies','requirements','location']
data = []
# Display and store away 2 pages of scrapped data from startupbeat.hkej.com
for i in range(1,5):
    quote_page = requests.get('https://hk.jobsdb.com/hk/jobs/accounting/'+str(i))
    sleep(randint(8,15))
    print("\n***** Page " + str(i) +" in action *****")
    soup = BeautifulSoup(quote_page.content,'html.parser')
    for article in soup.find_all("div", attrs={"class":"vlwSXhJ"}):
    # for article in soup.find_all('div',class_='archive-text'):
        page_no = str(i)
        job_title = article.span.text.strip()
        url = article.a.get('href')
        company = article.div.next_sibling.text.strip()
        requirements = article.ul.text.strip()
        location = soup.find("a",class_="_2nhUMny").text
        
        print(job_title)
        print(url)
        print(company)
        print(requirements)
        print(location)

        data.append((page_no, job_title, url, company, requirements, location))
df = pd.DataFrame(data,columns = header)
df.to_csv('accounting_jobs.csv', sep='\t', encoding='utf-8')
```
